```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

fake = Faker()

# CONFIG
NUM_CUSTOMERS = 150000
NUM_PRODUCTS = 8000
NUM_REPS = 1500
NUM_ROUTES = 500
NUM_ORDERS = 1500000

# ---------------- CUSTOMERS ----------------
customers = []
for i in range(NUM_CUSTOMERS):
    customers.append([
        i+1,
        fake.company(),
        random.choice(["Colombo","Galle","Kandy","Jaffna","Negombo"]),
        fake.phone_number() if random.random() > 0.1 else None,
        random.choice(["Active","Inactive",""]),
        fake.date_between(start_date='-3y', end_date='today')
    ])

df_customers = pd.DataFrame(customers, columns=[
    "CustomerID","CustomerName","City","Phone","Status","CreatedDate"
])

df_customers.to_csv("customers.csv", index=False)


# ---------------- PRODUCTS ----------------
products = []
for i in range(NUM_PRODUCTS):
    products.append([
        i+1,
        fake.word(),
        round(random.uniform(50, 10000), 2),
        random.choice(["Food","Beverage","Cosmetics","Household"])
    ])

df_products = pd.DataFrame(products, columns=[
    "ProductID","ProductName","Price","Category"
])

df_products.to_csv("products.csv", index=False)


# ---------------- SALES REPS ----------------
reps = []
for i in range(NUM_REPS):
    reps.append([
        i+1,
        fake.name(),
        random.choice(["North","South","West","Central"]),
        random.randint(1, NUM_ROUTES)
    ])

df_reps = pd.DataFrame(reps, columns=[
    "RepID","RepName","Region","RouteID"
])

df_reps.to_csv("reps.csv", index=False)


# ---------------- ROUTES ----------------
routes = []
for i in range(NUM_ROUTES):
    routes.append([
        i+1,
        f"Route-{i+1}",
        random.choice(["Urban","Rural"])
    ])

df_routes = pd.DataFrame(routes, columns=[
    "RouteID","RouteName","Type"
])

df_routes.to_csv("routes.csv", index=False)


# ---------------- ORDERS ----------------
orders = []
start_date = datetime(2022,1,1)

for i in range(NUM_ORDERS):
    orders.append([
        i+1,
        random.randint(1, NUM_CUSTOMERS),
        random.randint(1, NUM_REPS),
        start_date + timedelta(days=random.randint(0, 1200)),
        random.choice(["Completed","Pending","Cancelled"])
    ])

df_orders = pd.DataFrame(orders, columns=[
    "OrderID","CustomerID","RepID","OrderDate","Status"
])

df_orders.to_csv("orders.csv", index=False)


# ---------------- ORDER ITEMS ----------------
items = []
for i in range(NUM_ORDERS * 3):
    items.append([
        i+1,
        random.randint(1, NUM_ORDERS),
        random.randint(1, NUM_PRODUCTS),
        random.randint(1, 25),
        round(random.uniform(50, 10000), 2)
    ])

df_items = pd.DataFrame(items, columns=[
    "ItemID","OrderID","ProductID","Qty","Price"
])

df_items.to_csv("order_items.csv", index=False)


# ---------------- PAYMENTS ----------------
payments = []
for i in range(800000):
    payments.append([
        i+1,
        random.randint(1, NUM_ORDERS),
        round(random.uniform(500, 100000), 2),
        random.choice(["Cash","Card","Transfer"]),
        fake.date_this_year()
    ])

df_payments = pd.DataFrame(payments, columns=[
    "PaymentID","OrderID","Amount","Method","Date"
])

df_payments.to_csv("payments.csv", index=False)


# ---------------- RETURNS ----------------
returns = []
for i in range(200000):
    returns.append([
        i+1,
        random.randint(1, NUM_ORDERS),
        random.randint(1, NUM_PRODUCTS),
        random.randint(1, 5),
        fake.date_this_year()
    ])

df_returns = pd.DataFrame(returns, columns=[
    "ReturnID","OrderID","ProductID","Qty","ReturnDate"
])

df_returns.to_csv("returns.csv", index=False)

print("✅ Massive SFA dataset generated!")
```
